```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PIL import Image
```

```python
def show(arr, filename = None):

    plt.imshow(arr)
    
    if filename:
        
        plt.imsave('./segment_data/segments/' + filename + '.jpg', arr)
        plt.show()
        
    else:
        
        plt.show()
```

```python
def segment(arr, n, m, p=30):
    
    i_y = np.arange(arr.shape[0] // p) * p
    i_x = np.arange(arr.shape[1] // p) * p
    
    y = i_y[n]
    x = i_x[m]
    
    return arr[y:y+p, x:x+p, :]
```

```python
def dim(arr, p=30):
    
    i_y = np.arange(arr.shape[0] // p) * p
    i_x = np.arange(arr.shape[1] // p) * p
    
    return len(i_y), len(i_x)
```

```python
file = '5.jpg'

img = Image.open('./segment_data/' + file)
arr = np.array(img)

show(arr)
```

```python
p = 60
len_n, len_m = dim(arr, p=p)
```

```python
#segment

for n in range(len_n):
    
    for m in range(len_m):
        
        seg = segment(arr, n, m, p=p)
        fn = file.split('.')[0]+'_'+str(n)+'_'+str(m)
        print(fn)
        show(seg, fn)
```

```python
# create csv

fn_list = []

for n in range(len_n):
    
    for m in range(len_m):
        
        fn = file.split('.')[0]+'_'+str(n)+'_'+str(m)
        fn_list.append(fn)
```
