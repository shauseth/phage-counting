```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plate_path = 'plates/'
files = sorted(os.listdir(plate_path))
files.remove('.DS_Store')
files
```

```python
def segment(filename, p = 60):
    pre = filename.split('.')[0]
    img = Image.open(plate_path + filename)
    arr = np.array(img)
    x, y, z = arr.shape
    m = x // p
    n = y // p
    a = (x - (m * p)) // 2
    b = (y - (n * p)) // 2
    x_start, x_stop = a, x - a
    y_start, y_stop = b, y - b
    arr = arr[x_start:x_stop, y_start:y_stop, :]
    for i, sec in enumerate(np.split(arr, m, axis = 0)):
        for j, seg in enumerate(np.split(sec, n, axis = 1)):
            name = pre + '_' + str(i + 1) + '_' + str(j + 1)
            plt.imshow(seg)
            plt.imsave('segments/' + name + '.jpg', seg)
            plt.show()
```

```python
def show_img(filename):
    img = Image.open(plate_path + filename)
    arr = np.array(img)
    plt.imshow(arr)
    plt.show()
```

```python
filename = '5.jpg'
show_img(filename)
```
